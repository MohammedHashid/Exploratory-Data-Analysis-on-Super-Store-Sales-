```python
summary = pd.DataFrame({'Profit':df.Quantity[df.Profit > 0].sum(), 
                        'No-Profit':df.Quantity[df.Profit == 0].sum(), 
                        'Loss':df.Quantity[(df.Profit) < 0].sum()},
                       index={'Count'}).T
plt.title('Overall Sales Summary', fontsize=20)
summary.Count.plot.pie(autopct='%1.2f%%',figsize=(7,7), label='Percentage', 
                       textprops = {"fontsize":15}, shadow=True, explode=(0.08,0.05,0))
plt.show()
```

```python
state_P = pd.DataFrame(df.groupby('State').sum()).sort_values('Profit', ascending=False)
plt.figure(figsize=(18,7))
plt.title("States' position in overall Sales", fontsize=15)
sns.barplot(state_P.index, state_P.Profit)
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(), annot=True, cbar=False, annot_kws={'size':14})
plt.show()
```

```python
cat = np.round(df.groupby('Category').sum(), decimals=2).sort_values('Profit', ascending=False)
plt.title('Overall Profit - By Category', fontsize=14)
cat.Profit.plot.pie(autopct='%1.2f%%', figsize=(7,7), label='Profit Percentage', 
                    textprops = {"fontsize":15}, explode=(0, 0, 0.2), shadow=True)
plt.show()
```

```python
subc = np.round(df.groupby('Sub-Category').sum(), decimals=2).sort_values('Quantity', ascending=False)
plt.title('Overall Profit - By Sub-Category', fontsize=14)
subc.Quantity.plot.pie(autopct='%2.2f%%', figsize=(12,15), label='Sub-Category Percentage', 
                    textprops = {"fontsize":15}, shadow=True)
plt.show()
```

```python
furn = df[df.Category == 'Furniture'].groupby('Sub-Category').sum().sort_values('Profit', ascending=False).iloc[:, [-1]]
plt.figure(figsize=(10,4))
plt.title('Profit - Furniture ')
sns.barplot( furn.Profit, furn.index,)
plt.show()
```

```python
sub = np.round(df.groupby('Sub-Category').sum(), decimals=2).sort_values('Profit', ascending=False)
plt.figure(figsize=(10,6))
plt.title('Profit - by Sub-Category', fontsize=14)
sns.barplot(sub.Profit, sub.index)
plt.xticks(rotation=45)
plt.show()
```

```python
seg = np.round(df.groupby('Segment').sum(), decimals=2).sort_values('Profit', ascending=False)
plt.title('Overall Profit - By Segment', fontsize=14)
seg.Profit.plot.pie(autopct='%1.2f%%', figsize=(7,7), label='Profit Percentage', 
                    textprops = {"fontsize":15}, explode=(0, 0, 0.2), shadow=True)
plt.show()
```

```python
sns.pairplot(data=df.iloc[:,-3:], kind='reg')
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].set_title('Discount vs Profit', fontsize=15)
ax[1].set_title('Quantity vs Profit', fontsize=15)
sns.lineplot(df.Discount, df.Profit, color='green', label='Profit Change', ax=ax[0])
sns.lineplot(df.Quantity, df.Profit, color='red', label='Profit Change', ax=ax[1])
plt.show()
```
